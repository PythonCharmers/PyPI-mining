```python
import requests
import re
import tempfile
import functools
import gzip
import json
import pickle
from collections import defaultdict
from itertools import islice
import pandas as pd
```

```python
URL = 'http://lidlbit.com:8888/'


def get_file_list(url, ext='gz'):
    """ 
    Get a file list from a http server. 
    """

    index = requests.get(url,  auth=('mining', 'pypi'))
    pattern = '<a href="(.*)">'.format(ext)
    return [ x for x in re.findall(pattern, index.content.decode()) if x.endswith(ext)]


@functools.lru_cache()
def get_file(url, filename):
    """
    Retrieve a file, into a temporary file.
    """
    
    data = requests.get('{}/{}'.format(url, filename), auth=('mining', 'pypi'))  
    temp = tempfile.NamedTemporaryFile()
    temp.write(data.content)
    temp.seek(0)
    return temp


def decompress_json(jsonfile):
    """
    Decompress the json named temporary file.
    """

    jsonfile.seek(0)
    return gzip.decompress(jsonfile.read())
    
```

```python
files = get_file_list(URL)
```

```python
from itertools import chain

def version_df(jsondata_date, pyver='python_three'):
    packages = []
    versions = []
    three_point_x = []

    for classifier in jsondata_date[pyver]:
        for row in jsondata_date[pyver][classifier]:
            try:
                version = row['version']
                package = row['package']
            except TypeError:
                package, version = row
            packages.append(package)
            versions.append(version)
            three_point_x.append(defaultdict(bool))
            three_point_x[-1][classifier] = True

    classifier_cols = defaultdict(list)
    for classifier in jsondata_date[pyver]:
        for row in three_point_x:
            classifier_cols[classifier].append(row[classifier])

    cols12 = {'package': packages, 'version': versions}
    allcols = dict(chain(cols12.items(), classifier_cols.items()))
    df = pd.DataFrame(allcols)
    return df.groupby(['package', 'version']).sum()
```

```python
import sys, time
try:
    from IPython.display import clear_output
    have_ipython = True
except ImportError:
    have_ipython = False

class ProgressBar:
    def __init__(self, iterations):
        self.iterations = iterations
        self.prog_bar = '[]'
        self.fill_char = '*'
        self.width = 40
        self.__update_amount(0)
        if have_ipython:
            self.animate = self.animate_ipython
        else:
            self.animate = self.animate_noipython

    def animate_ipython(self, iter):
        print('\r', self, end=None)
        sys.stdout.flush()
        self.update_iteration(iter + 1)

    def update_iteration(self, elapsed_iter):
        self.__update_amount((elapsed_iter / float(self.iterations)) * 100.0)
        self.prog_bar += '  %d of %s complete' % (elapsed_iter, self.iterations)

    def __update_amount(self, new_amount):
        percent_done = int(round((new_amount / 100.0) * 100.0))
        all_full = self.width - 2
        num_hashes = int(round((percent_done / 100.0) * all_full))
        self.prog_bar = '[' + self.fill_char * num_hashes + ' ' * (all_full - num_hashes) + ']'
        pct_place = (len(self.prog_bar) // 2) - len(str(percent_done))
        pct_string = '%d%%' % percent_done
        self.prog_bar = self.prog_bar[0:pct_place] + \
            (pct_string + self.prog_bar[pct_place + len(pct_string):])

    def __str__(self):
        return str(self.prog_bar)
```

```python
def collect_data(files, pyver='python_three', progressbar=False):
    if progressbar:
        p = ProgressBar(len(files))

    numbers = []
    for i in range(0, len(files), 30):
        if i % 10 == 0 and progressbar:
            p.animate(i)
        jsondata = json.loads(decompress_json(get_file(URL, files[i])).decode())
        dates = jsondata.keys()
        assert len(dates) == 1
        date = list(dates)[0]

        py_k = jsondata[date][pyver]
        classifiers_k = sorted(py_k.keys())

        df = version_df(jsondata[date], pyver=pyver)
        numbers.append(df.sum())

        with open('numbers_%s.pickle' % pyver, 'wb') as f:
            pickle.dump(numbers, f)
    return numbers
```
