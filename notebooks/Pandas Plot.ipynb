```python
import dateutil
import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
frame_py2 = pd.read_json('/Users/nfaggian/Desktop/py2.json')
frame_py3 = pd.read_json('/Users/nfaggian/Desktop/py3.json')
```

```python
py3_series = frame_py3['Programming Language :: Python :: 3'].apply(len)
py2_series = frame_py2['Programming Language :: Python :: 2'].apply(len)
```

```python
for freq in ['1h', '12h', '1d']:

    fig, ax = plt.subplots(figsize=(8,6))
    fig.suptitle('Package growth ({})'.format(freq))
    (py3_series.resample(freq, how='mean') - py3_series[0]).plot(ax=ax, style='r.-')
    (py2_series.resample(freq, how='mean') - py2_series[0]).plot(ax=ax, style='b.-')
    ax.legend(['Python (3)', 'Python (2)'], loc='upper left')
    sns.set_style("white")
```
