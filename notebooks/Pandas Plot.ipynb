```python
import pymongo
```

```python
import dateutil
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
conn = pymongo.MongoClient('192.168.0.5', 27017)
```

```python
pypi = conn.pypi
```

```python
py3_data = {}
py2_data = {}

for record in pypi.python_stats.find():
    date = [x for x in record.keys() if x != '_id'][0]
    py3_data[dateutil.parser.parse(date)] = {k:len(v) for k, v in record[date]['python_three'].items()}
    py2_data[dateutil.parser.parse(date)] = {k:len(v) for k, v in record[date]['python_two'].items()}
        
```

```python
df3 = pd.DataFrame(py3_data).T
py3_series = df3['Programming Language :: Python :: 3']

df2 = pd.DataFrame(py2_data).T
py2_series = df2['Programming Language :: Python :: 2']
```

```python
for freq in ['1h', '12h', '1d']:

    fig, ax = plt.subplots(figsize=(8,6))
    fig.suptitle('Package growth ({})'.format(freq))
    (py3_series.resample(freq, how='mean') - py3_series[0]).plot(ax=ax, style='r.-')
    (py2_series.resample(freq, how='mean') - py2_series[0]).plot(ax=ax, style='b.-')
    ax.legend(['Python (3)', 'Python (2)'], loc='upper left')
    sns.set_style("white")
```
